
## C2) Mini-project: clean a small “survey” dataset
Create a list of dictionaries like this (you can use these exact examples):
- `{ "name": "Ali", "age": "19", "city": "Rabat", "score": "14" }`
- `{ "name": "Mina", "age": "", "city": "Casablanca", "score": "17" }`
- `{ "name": "Youssef", "age": "22", "city": "Rabat", "score": "" }`
- `{ "name": "Sara", "age": "21", "city": "Marrakech", "score": "15" }`




```python
# Survey data
survey = [
    { "name": "Ali", "age": "19", "city": "Rabat", "score": "14" },
    { "name": "Mina", "age": "", "city": "Casablanca", "score": "17" },
    { "name": "Youssef", "age": "22", "city": "Rabat", "score": "" },
    { "name": "Sara", "age": "21", "city": "Marrakech", "score": "15" }
]

```

**Tasks:**
1. Convert `age` and `score` to numbers when possible.
2. Replace missing numeric values (`""`) with `None`.



```python
# Clean data: convert age/score, handle missing values
for entry in survey:
    entry["age"] = int(entry["age"]) if entry["age"] else None
    entry["score"] = int(entry["score"]) if entry["score"] else None

print(survey)
```

3. Compute: Average score (ignore missing scores)

```python

# Compute average score (ignore None)
scores = [e["score"] for e in survey if e["score"] is not None]
avg_score = sum(scores) / len(scores) if scores else None
print("\nAverage score:", avg_score)
```

4. Compute  Count of students per city (use a dict)


```python
# Count students per city
city_counts = {}
for entry in survey:
    city = entry["city"]
    city_counts[city] = city_counts.get(city, 0) + 1

print("Students per city:", city_counts)
```

3. Compute  Average score (ignore missing scores)

```python
# Clean data: convert age/score, handle missing values
for entry in survey:
    entry["age"] = int(entry["age"]) if entry["age"] else None
    entry["score"] = int(entry["score"]) if entry["score"] else None
```

5. Print a clean summary report.

```python

# Print summary
print("Cleaned survey data:")
for entry in survey:
    print(entry)

print("\nAverage score:", avg_score)
print("Students per city:", city_counts)
```

## Pythonic options (Additional)
Short patterns you will see a lot in real code (especially for data work).

#### 1) f-strings (clean formatting)


```python
name = "Sara"
score = 15
print(f"{name} scored {score}/20")
```

#### 2) enumerate (index + value)


```python

cities = ["Rabat", "Casablanca", "Marrakech"]
for i, city in enumerate(cities, start=1):
    print(i, city)

```

#### 3) list comprehension (build a new list)


```python
scores = [12, 9, 20, 15]
passed = [s for s in scores if s >= 10]
print("passed:", passed)
```

#### 4) dict.get (safe lookup with default)


```python

student = {"name": "Ali"}
print("major:", student.get("major", "Unknown"))
```

### # 5) zip (loop in parallel)


```python

names = ["Ali", "Mina", "Sara"]
ages = [19, 20, 21]
for n, a in zip(names, ages):
    print(n, "->", a)
```
