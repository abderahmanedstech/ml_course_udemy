## Mini-checkpoint
- In 2–3 lines, explain the difference between `list`, `tuple`, `dict`, and `set`.

- A `list` is an ordered, mutable collection that allows duplicate elements.  
- A `tuple` is similar to a list but is immutable (cannot be changed after creation).  
- A `dict` stores key-value pairs for fast lookup by key, while a `set` is an unordered collection of unique elements (no duplicates allowed).
