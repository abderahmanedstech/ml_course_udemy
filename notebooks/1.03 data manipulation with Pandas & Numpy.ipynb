# 1.03 — Data manipulation with NumPy & pandas

## What’s the difference?
**NumPy**
- Core object: **`ndarray`** (n-dimensional array)
- Best for: fast numerical computing, vectors/matrices, linear algebra, broadcasting
- Think: *math with arrays*

**pandas**
- Core objects: **`Series`** (1D) and **`DataFrame`** (2D table)
- Best for: tabular data, column names, missing values, joins/merges, groupby/aggregation
- Think: *Excel-like tables + data cleaning*

![pandas vs numpy](images/numpy_pandas.png)

## Quick rule of thumb
- Use **NumPy** when your data is mostly numeric and you want fast array math.
- Use **pandas** when your data has columns (possibly mixed types) and you need filtering, grouping, merging, or cleaning.

## NumPy ↔ pandas (how they work together)
- A pandas `DataFrame` stores columns as NumPy-like arrays internally.
- Use **pandas** for labels + missing values + table operations.
- Use **NumPy** for fast numeric math once you have clean numeric arrays.
