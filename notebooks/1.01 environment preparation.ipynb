# 1) What is Anaconda? (and how is it different from Python?)

**Python** is a programming language (the interpreter + standard library).

**Anaconda** is a *distribution* that bundles:
- Python ( with associated package also **pip**)
- A package manager + environment manager called **conda**
- Many popular data/ML packages (NumPy, pandas, etc.)
- Optional tools like **Anaconda Navigator** (GUI)

**Key difference (simple):**
- Python = the language/runtime.
- Anaconda = Python + conda + a curated ecosystem to install/manage scientific packages and isolated environments more easily.

# 2) Download + install Anaconda (Windows)

**Download (official):**
- https://www.anaconda.com/download

**Install steps (recommended defaults):**
1. Download the Windows installer for your system (usually 64‑bit).
2. Run the installer.
3. Choose **Just Me** (recommended).
4. Pick an install location (default is fine).
5. When asked about PATH, keep the default recommendations (usually *do not* add Anaconda to PATH).
6. Finish install.
7. Open **Anaconda Prompt** from the Start Menu to use conda.

**Quick verification:**
- In **Anaconda Prompt** run:
  - `conda --version`
  - `python --version`

# 3) Work with virtual environments (best practice)

Using environments keeps each course/project independent (no package conflicts).

## Conda environments (recommended with Anaconda)

**Create an environment:**
- `conda create -n mlcourse python=3.11`

**Activate it:**
- `conda activate mlcourse`

**Install packages into that environment:**
- `conda install numpy pandas scikit-learn`
- or (inside the env): `pip install matplotlib seaborn`

**List / remove envs:**
- `conda env list`
- `conda remove -n mlcourse --all`

![Conda venv](images/conda.png)

**Another Option to install package is pip of  Python**

**Install:**
- `pip install -U pip`
- `pip install jupyterlab numpy pandas`

**Uninstall:**
- `pip uninstall <package_name>`



> For this course, we’ll mainly use **conda envs** since you are installing Anaconda.



# 4) Launch Jupyter (prefer JupyterLab in the browser)

First, activate the environment you want to use:
- `conda activate mlcourse`

## Install JupyterLab (once per environment)
- `pip install jupyterlab` 
  - or: `conda install -c conda-forge jupyterlab`

## Launch JupyterLab
- `jupyter lab`

This will open a browser tab. If it doesn’t open automatically, copy/paste the URL shown in the terminal (it includes a token).

## (Optional) Launch classic Jupyter Notebook
- `jupyter notebook`

# 5) Use VS Code with your environment + Jupyter

VS Code can run notebooks and Python scripts using the same conda environment.

## Step 1 — Install VS Code extensions
In VS Code, install:
- **Python** (Microsoft)
- **Jupyter** (Microsoft)

## Step 2 — Select the right Python interpreter
1. Open the Command Palette: `Ctrl+Shift+P`
2. Choose **Python: Select Interpreter**
3. Pick the interpreter from your conda env (e.g. something like `...\anaconda3\envs\mlcourse\python.exe`).

## Step 3 — Work with notebooks in VS Code
- Open an `.ipynb` file and run cells with the ▶ button.
- When prompted for a kernel, select the **mlcourse** environment.

## If the kernel is missing
Activate your env in Anaconda Prompt, then run:
- `conda activate mlcourse`
- `conda install ipykernel`
- `python -m ipykernel install --user --name mlcourse --display-name "Python (mlcourse)"`

# Cloud option (if your laptop resources are limited) — Google Colab

If you want to run notebooks in the cloud (CPU/GPU) without installing anything locally, you can use **Google Colab**. It’s great for **small exercises** and demos; resources and session time are limited on the free tier.

**Open Colab:**
- https://colab.research.google.com/
- Click **File → New notebook** (or **Upload notebook** to upload your `.ipynb`).

**Choose hardware (optional):**
- **Runtime → Change runtime type** → choose **CPU / GPU / TPU** (GPU may be limited/queued).

**Install extra packages (inside a code cell):**
- `!pip install numpy pandas scikit-learn`

**Save / load files:**
- You can upload files via the left panel (Files)
- Or mount Google Drive:
  - `from google.colab import drive`
  - `drive.mount('/content/drive')`

**Important limitations (good to know):**
- Sessions can disconnect; work is not guaranteed to persist unless saved to Drive/GitHub.
- Free resources are limited, so use it mainly for learning labs and small datasets.

**Other cloud notebook options (similar idea):**
- Kaggle Notebooks (free, good for small projects/datasets)
- Any JupyterLab server (university/work server or cloud VM) if you have access
