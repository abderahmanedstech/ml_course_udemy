# 1.03 — Data manipulation with NumPy & pandas

## What’s the difference?
**NumPy**
- Core object: **`ndarray`** (n-dimensional array)
- Best for: fast numerical computing, vectors/matrices, linear algebra, broadcasting
- Think: *math with arrays*

**pandas**
- Core objects: **`Series`** (1D) and **`DataFrame`** (2D table)
- Best for: tabular data, column names, missing values, joins/merges, groupby/aggregation
- Think: *Excel-like tables + data cleaning*

![pandas vs numpy](images/numpy_pandas.png)

## Quick rule of thumb
- Use **NumPy** when your data is mostly numeric and you want fast array math.
- Use **pandas** when your data has columns (possibly mixed types) and you need filtering, grouping, merging, or cleaning.

## NumPy demo (arrays = fast math)
We’ll create arrays, slice them, and apply vectorized operations (no Python loops).

#### 1. Create arrays


```python
list_a = [1, 2, 3, 4, 5]
list_b = [10, 20, 30, 40, 50]
```

```python
list_a + list_b
```

```python
a = np.array(list_a)
b = np.array(list_b)
print('a:', a)
print('b:', b)
```

#### 2. Vectorized operations (element-wise)


```python

print('a + b:', a + b)
print('a * 2:', a * 2)
```

#### 3. Boolean filtering


```python

mask = a >= 3
print('mask:', mask)
print('a[a>=3]:', a[mask])
```

#### 4. 2D array + slicing


```python

m = np.array([[1, 2, 3], [4, 5, 6]])
print('m:\n', m)
print('first row:', m[0, :])
print('second column:', m[:, 1])
```

## pandas demo (tables = labeled columns)
#### 1.Create a small DataFrame


```python
df_small = pd.DataFrame({
    'name': ['John', 'Emily', 'Michael', 'Sophia'],
    'city': ['New York', 'Los Angeles', 'Chicago', 'Houston'],
    'score': [14, 17, 15, 9],
})
df_small
```

#### 2. Select columns / rows


```python
print(df_small[['name', 'score']])

```

#### 3. print score high to 15

```python
high_score = df_small[df_small['score'] >= 15]
print('\nscore >= 15')
print(high_score)

```

#### 4. Create a new column

```python
df_small['passed'] = df_small['score'] >= 10
print('\nwith passed column')
print(df_small)

```

```python
# Groupby: average score per city
avg_by_city = df_small.groupby('city', as_index=False)['score'].mean()
print('\naverage score by city')
print(avg_by_city)
```

## NumPy ↔ pandas (how they work together)
- A pandas `DataFrame` stores columns as NumPy-like arrays internally.
- Use **pandas** for labels + missing values + table operations.
- Use **NumPy** for fast numeric math once you have clean numeric arrays.

```python
# Convert a DataFrame column to NumPy, do math, and put it back
scores_np = data_frame['math_score'].to_numpy()
print('scores_np type:', type(scores_np), 'values:', scores_np)

```

```python
z = (scores_np - scores_np.mean()) / scores_np.std()  # simple normalization
data_frame['math_score_z'] = z
data_frame.head(5)
```
