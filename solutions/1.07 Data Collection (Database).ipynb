# LAB 1.07 Data Collection (Database)


### Import Required Libraries

```python
import sqlite3
import pandas as pd
```

### Create a Sample Database (For Students)

```python
# Create SQLite database
conn = sqlite3.connect("data/ml_course.db")
cursor = conn.cursor()

# Create table
cursor.execute("""
CREATE TABLE IF NOT EXISTS customers (
    customer_id INTEGER PRIMARY KEY,
    age INTEGER,
    income REAL,
    city TEXT,
    purchased INTEGER
)
""")

# Insert sample data
data = [
    (25, 30000, "Paris", 0),
    (35, 55000, "London", 1),
    (45, 70000, "Berlin", 1),
    (23, 28000, "Madrid", 0),
    (50, 90000, "Rome", 1)
]

cursor.executemany(
    "INSERT INTO customers (age, income, city, purchased) VALUES (?, ?, ?, ?)",
    data
)

conn.commit()
```

### Read Data from Database (SQL → pandas)

```python
query = "SELECT * FROM customers"
df = pd.read_sql(query, conn)

df
```

### Filtering Data Using SQL

```python
query = """
SELECT age, income, purchased
FROM customers
WHERE income > 40000
"""

df_filtered = pd.read_sql(query, conn)
df_filtered
```
